```python
#play this code only once
hd['temp'] = (hd['temp'] * (39 -(-8))+(-8)).round(1)
hd['atemp'] = (hd['atemp'] * (50 + 16)-16).round(1)
hd['hum'] = (hd['hum'] * 100).round(1)
hd['windspeed'] = (hd['windspeed'] * 67 * 1.60934).round(1)
hd["dteday"] = pd.to_datetime(hd['dteday']) #to datetime
hd = hd.rename(columns={"dteday": "Date"})

dd['temp'] = (dd['temp'] * (39 + 8)-8).round(1)
dd['atemp'] = (dd['atemp'] * (50 + 16)-16).round(1)
dd['hum'] = (dd['hum'] * 100).round(1)
dd['windspeed'] = (dd['windspeed'] * 67 * 1.60934).round(1) #from mph to km/h
dd["dteday"] = pd.to_datetime(dd['dteday'])
dd = dd.rename(columns={"dteday": "Date"})
```

```python
#creating new weekday column to make it same with the database 2021-23 and putting to the same position where it was.
#now 0 is Monday, 6 is Sunday
dd.drop('weekday', axis=1, inplace=True)
dd['weekday'] = dd['Date'].dt.weekday
weekday_column = dd.pop('weekday')
new_position = dd.columns.get_loc("holiday") + 1
dd.insert(new_position, 'weekday', weekday_column)

hd.drop('weekday', axis=1, inplace=True)
hd['weekday'] = hd['Date'].dt.weekday
weekday_column = hd.pop('weekday')
new_position = hd.columns.get_loc("holiday") + 1
hd.insert(new_position, 'weekday', weekday_column)
```
