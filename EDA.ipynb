```python
dd['year'] = dd['yr'].apply(lambda value: '2011' if value == 0 else ('2012' if value == 1 else ('2021' if value == 2 else ('2022' if value == 3 else '2023'))))
hd['year'] = hd['yr'].apply(lambda value: '2011' if value == 0 else ('2012' if value == 1 else ('2021' if value == 2 else ('2022' if value == 3 else '2023'))))
```

```python
dd["Date"] = pd.to_datetime(dd['Date'])
hd["Date"] = pd.to_datetime(hd['Date'])
```

```python
dd.drop(columns=['yr', 'instant'], inplace=True)
hd.drop(columns=['yr', 'instant'], inplace=True)
```

```python
hd['Casual'] = hd['Casual'].astype(int)
hd['Member'] = hd['Member'].astype(int)
hd['Total_rides'] = hd['Total_rides'].astype(int)
```

```python
dd.to_csv('dd_11-23.csv', index=False)
hd.to_csv('hd_11-23.csv', index=False)
```
