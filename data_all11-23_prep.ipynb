```python
df11['Start date'] = pd.to_datetime(df11['Start date'])
df11['End date'] = pd.to_datetime(df11['End date'])
df11['Time'] = pd.to_timedelta(df11['Time'])
df11['Member type'] = df11['Member type'].astype('category')
df11['Member type'] = df11['Member type'].str.lower()
```

```python
df21['Start date'] = pd.to_datetime(df21['started_at'])
df21['End date'] = pd.to_datetime(df21['ended_at'])
df21 = df21.rename(columns={"start_station_id": "Start station number", 
                            "end_station_name": "End station", 
                            "start_station_name": "Start station",
                            "end_station_id": "End station number", 
                            'member_casual': "Member type"})
df21['Member type'] = df21['Member type'].astype('category')
df21 = df21.drop(columns=['started_at', 'ended_at', 'ride_id'])
df21['Time'] = df21['End date'] - df21['Start date']
df21['Duration'] = df21['Time'].dt.total_seconds().astype('int64')
```

```python
df21 = df21[df21['rideable_type'] == 'classic_bike']
df21 = df21.drop(columns=['rideable_type'])
```

```python
df11 = df11.merge(geo[['start station', 'start_lat', 'start_lng']], how='left', left_on='Start station', right_on='start station')
```
